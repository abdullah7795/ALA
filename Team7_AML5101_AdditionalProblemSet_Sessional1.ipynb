**Problem 3.1**: In order to measure how rough or wiggly a signal represented by a vector is, we measure the so called *Dirichlet energy* of the signal which is defined as:

$$(x_2-x_1)^2+(x_3-x_2)^2+\cdots+(x_{100}-x_{99})^2.$$

Calculate the Dirichlet energy of the vector $x$ as a matrix-vector product for the three different frequency values. How do the values compare to each other? Use the following clue:

$\begin{align*}\text{Dircihlet energy of }x &=\lVert Dx\lVert \\&=\left\lVert\begin{bmatrix}-1 & ? & ? &\cdots&?&?\\0 & ? & 1 &\cdots&?&?\\? & ? & ? &\cdots&?&?\\ &  & \vdots &\vdots&&\\? & ? & ? &\cdots&-1&?\end{bmatrix}\begin{bmatrix}x_1\\x_2\\x_3\\\vdots\\x_{99}\\x_{100}\end{bmatrix}\right\rVert.\end{align*}$

```
# Build the matrix D
n = len(x) 
a = -1*np.ones(n-1)
b = 1*np.ones(n)
D = np.diag(a, -1) + np.diag(b)
D = D[1:, :]
print(D)
# Calculate Dirichlet energy
np.linalg.norm(np.dot(D, x))
```
