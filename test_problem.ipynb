```python
import numpy as np
```

```python
#original vector
A=[3,4,5,5,6,7]
B=[2,2,2,8,8,8]
C=[15,15,15,15,15,15]
```

```python
#deviations vector
print(A-np.mean(A))
print(B-np.mean(B))
print(C-np.mean(C))
```

```python
#squared deviations vector
print((A-np.mean(A))**2)
print((B-np.mean(B))**2)
print((C-np.mean(C))**2)
```

```python
#(4) average of the squared deviations vector (this is the variance) 
print(np.average((A-np.mean(A))**2))
print(np.average((B-np.mean(B))**2))
print(np.average((C-np.mean(C))**2))
```

```python
#(5) square root of the variance (this is the standard deviation).
print(np.sqrt(np.average((A-np.mean(A))**2)))
print(np.sqrt(np.average((B-np.mean(B))**2)))
print(np.sqrt(np.average((C-np.mean(C))**2)))
```
