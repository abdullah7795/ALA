```python
import sympy as sp
# Augmented matrix
A = sp.Matrix([[1,2,-1,-1,1], [2,4,-2,3,3], [-1,1,-2,4,2]])
# Return the rref and the pivot column indices as a tuple
print(A.rref())
```
